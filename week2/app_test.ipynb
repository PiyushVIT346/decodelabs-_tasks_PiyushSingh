```python
import os
from google import genai
from google.genai import types

# Initialize the client.
# Ensure you have set your API key in your environment variables:
os.environ["GEMINI_API_KEY"] = "your_gemini_api_key"
client = genai.Client()
```

```python
import time

def generate_marketing_copy(product_name, description, platform, tone, temperature=0.7, top_p=0.9, max_retries=3):
    """
    Takes raw product details and generates tailored marketing copy, 
    with built-in retry logic for handling temporary server overloads (503 errors).
    """
    # 1. Dynamic Prompt Template Compilation
    prompt = (
        "You are an expert digital marketer and copywriter.\n"
        "Your task is to transform the following raw product information into highly engaging copy.\n\n"
        "PRODUCT DETAILS:\n"
        f"- Product Name: {product_name}\n"
        f"- Raw Description: {description}\n\n"
        "TARGET PLATFORM:\n"
        f"{platform}\n\n"
        "REQUIRED TONE:\n"
        f"{tone}\n\n"
        "INSTRUCTIONS:\n"
        f"Write the final marketing copy for {platform}. Embody the {tone} tone completely. "
        "Keep the formatting appropriate for the platform. Do not use emojis in the output."
    )

    # 2. Inference Parameter Tuning
    config = types.GenerateContentConfig(
        temperature=temperature,
        top_p=top_p
    )

    # 3. Model Execution with Retry Logic
    for attempt in range(max_retries):
        try:
            response = client.models.generate_content(
                model='gemini-2.5-flash',
                contents=prompt,
                config=config
            )
            return response.text
            
        except Exception as e:
            # Check if it's a server overload/503 error
            if "503" in str(e) or "UNAVAILABLE" in str(e) or "429" in str(e):
                if attempt < max_retries - 1:
                    wait_time = 2 ** attempt  # Exponential backoff: waits 1s, then 2s, then 4s
                    print(f"API busy. Retrying {platform} generation in {wait_time} seconds (Attempt {attempt + 1} of {max_retries - 1})...")
                    time.sleep(wait_time)
                else:
                    print(f"Failed to generate copy for {platform} after {max_retries} attempts.")
                    raise e
            else:
                # If it's a different kind of error (like an auth error), raise it immediately
                raise e
```

```python
product = "AeroDesk Pro"
raw_desc = "A standing desk with a dual-motor lift system, built-in wireless charging pad, and programmable height presets. Made from sustainable bamboo."

print("==============================")
print(" LINKEDIN POST ")
print("==============================")
# LinkedIn: Professional tone, lower temperature (0.4) for focused, factual, and highly logical delivery
linkedin_copy = generate_marketing_copy(
    product_name=product,
    description=raw_desc,
    platform="LinkedIn",
    tone="Professional, thought-leadership, focused on productivity, workplace ergonomics, and sustainability",
    temperature=0.4,
    top_p=0.8
)
print(linkedin_copy)
```

```python
print("==============================")
print(" INSTAGRAM POST ")
print("==============================")
# Instagram: Punchy tone, higher temperature (0.8) to encourage creative variance and lifestyle-oriented phrasing
ig_copy = generate_marketing_copy(
    product_name=product,
    description=raw_desc,
    platform="Instagram",
    tone="Trendy, energetic, lifestyle-focused, purely text-based aesthetic",
    temperature=0.8,
    top_p=0.95
)
print(ig_copy)
print("\n\n")
```

```python
print("==============================")
print(" EMAIL CAMPAIGN ")
print("==============================")
# Email: Persuasive, direct-response, balanced temperature (0.6) for optimal readability and conversion
email_copy = generate_marketing_copy(
    product_name=product,
    description=raw_desc,
    platform="Email Newsletter",
    tone="Persuasive, addressing pain points of sitting all day, clear call-to-action, urgent but not spammy",
    temperature=0.6,
    top_p=0.9
)
print(email_copy)
```
