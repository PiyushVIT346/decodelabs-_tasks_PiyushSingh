```python
import os
import random
import requests
from PIL import Image
import ipywidgets as widgets
from IPython.display import display
from requests.adapters import HTTPAdapter
from urllib3.util import Retry

API_URL = "https://image.pollinations.ai/prompt/"

print("Core studio environment and network dependencies initialized successfully.")
```

```python
def generate_and_stream_image_studio(
    prompt: str, 
    aspect_ratio: str, 
    resolution_scale: str, 
    steps: int, 
    seed: int, 
    output_filename: str = "studio_artwork.png"
):
    """
    Executes the 6-stage blueprint framework:
    1. Formulates parameterized options payload and scales dimensions.
    2. Implements connection pooling with automatic backoff retries.
    3. Leverages browser-agent masquerading to prevent CDN packet drops.
    4. Pipes response using a memory-safe 64KiB chunk loop directly to disk storage.
    5. Forces pixel decoding matrix integrity checks.
    """
    # Stage 1: Dynamic Parameter Dimension Mapper
    base_dim = 1024 if resolution_scale == "High (1024p)" else 768
    if aspect_ratio == "1:1 Square":
        width, height = base_dim, base_dim
    elif aspect_ratio == "16:9 Landscape":
        width, height = base_dim, int(base_dim * (9 / 16))
    elif aspect_ratio == "9:16 Portrait":
        width, height = int(base_dim * (9 / 16)), base_dim
    elif aspect_ratio == "4:3 Classic":
        width, height = base_dim, int(base_dim * (3 / 4))
    
    chosen_seed = seed if seed != -1 else random.randint(1, 99999999)
    print(f"\n[Stage 1] Formulation Payload Applied: Shape Constraint {width}x{height} | Seed: {chosen_seed}")
    
    # URL encode parameters directly into the high-availability GET routing track
    sanitized_prompt = requests.utils.quote(prompt)
    target_url = f"{API_URL}{sanitized_prompt}?width={width}&height={height}&seed={chosen_seed}&nologo=true"
    
    # Stage 2 & 3: Network API Gateway Integration with Masqueraded Headers
    print("[Stage 2 & 3] Routing payload safely via resilient security gateways...")
    
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36",
        "Accept": "image/avif,image/webp,image/apng,image/svg+xml,image/*,*/*;q=0.8"
    }
    
    # Configure resilient retry limits for unstable or busy network channels
    retries = Retry(
        total=3, 
        backoff_factor=1, # Exponential delay backoff: 1s, 2s, 4s
        status_forcelist=[500, 502, 503, 504],
        raise_on_status=False
    )
    
    session = requests.Session()
    session.mount("https://", HTTPAdapter(max_retries=retries))
    
    try:
        # Request stream connection layout with a 5s connection / 60s read timeout policy
        response = session.get(target_url, headers=headers, stream=True, timeout=(5.0, 60))
        response.raise_for_status()
        
    except requests.exceptions.ConnectionError as conn_err:
        print(f"[Gateway Error] Connection pool refused by the endpoint.\nDetails: {conn_err}")
        return None
    except requests.exceptions.Timeout:
        print("[Gateway Error] Core network gateway timeout constraint exceeded.")
        return None
    except Exception as api_err:
        print(f"[API Error] Critical gateway failure exception: {api_err}")
        return None

    # Stage 4 & 5: Memory-Safe Transport Protocol and Integrity Check
    print("[Stage 4] Streaming payload directly to local disk via memory-safe 64KiB chunk pipeline...")
    try:
        # Safely remove existing artifacts to ensure clean writing streams
        if os.path.exists(output_filename):
            os.remove(output_filename)
            
        with open(output_filename, 'wb') as f:
            for chunk in response.iter_content(chunk_size=65536): # Mapped to specific 64KiB architecture thresholds
                if chunk:
                    f.write(chunk)
                    
        # Check that actual byte payload sizes have crossed threshold arrays
        file_size = os.path.getsize(output_filename)
        if file_size < 1000:
            print(f"[Storage Error] Stream error: file payload too small ({file_size} bytes). Endpoint might be serving text errors.")
            return None
            
        print(f"[Stage 5] Transport written safely ({file_size} bytes). Verifying canvas pixel decoding matrix...")
        
        # Cold execution pixel load to verify the visual asset pipeline integrity
        with Image.open(output_filename) as verified_img:
            verified_img.load()
            print("[Stage 5 Verification Passed] Image payload is completely stable and renderable.")
            return verified_img
            
    except Exception as stream_err:
        print(f"[Storage/Verification Error] Asset generation failed downstream checks: {stream_err}")
    return None
```

```python
# 1. Define interactive UI components layout
prompt_widget = widgets.Textarea(
    value='A highly detailed, cinematic digital rendering of a futuristic laboratory inside a greenhouse, cyberpunk elements',
    placeholder='Type your design prompt instructions...',
    description='Prompt:',
    layout=widgets.Layout(width='90%', height='80px')
)

ratio_widget = widgets.Dropdown(
    options=['1:1 Square', '16:9 Landscape', '9:16 Portrait', '4:3 Classic'],
    value='16:9 Landscape',
    description='Aspect Ratio:',
)

scale_widget = widgets.Dropdown(
    options=['Standard (768p)', 'High (1024p)'],
    value='Standard (768p)',
    description='Res Quality:',
)

steps_widget = widgets.IntSlider(
    value=4,
    min=1,
    max=12,
    step=1,
    description='Steps:',
)

seed_widget = widgets.IntText(
    value=-1,
    description='Seed (-1 Rand):',
)

generate_btn = widgets.Button(
    description='Orchestrate Studio Generation',
    button_style='info',
    tooltip='Click to safely call pipeline and process input configuration options',
    icon='paint-brush',
    layout=widgets.Layout(width='40%', height='40px')
)

# Container target area specifically for updating/displaying the image output cleanly
output_area = widgets.Output()

# 2. Bind execution pipeline directly inside button event loop
def on_click_orchestrate(b):
    with output_area:
        output_area.clear_output() # Clears output context windows completely
        
        # Capture current widget state selections
        chosen_prompt = prompt_widget.value
        chosen_ratio = ratio_widget.value
        chosen_scale = scale_widget.value
        chosen_steps = steps_widget.value
        chosen_seed = seed_widget.value
        
        print("Launching visual orchestrator studio generation pipeline...")
        print(f"User Prompt Input: '{chosen_prompt}'")
        
        generate_btn.disabled = True
        try:
            # Calls the verified 64KiB chunk streaming function
            img_asset = generate_and_stream_image_studio(
                prompt=chosen_prompt,
                aspect_ratio=chosen_ratio,
                resolution_scale=chosen_scale,
                steps=chosen_steps,
                seed=chosen_seed,
                output_filename="studio_artwork.png"
            )
            
            if img_asset is not None:
                print("\nRendering visual asset canvas matrix below:")
                display(img_asset)
                
        except Exception as pipeline_err:
            print(f"\n[Execution Failure] Runtime error: {pipeline_err}")
        finally:
            generate_btn.disabled = False

# Connect action observer mappings
generate_btn.on_click(on_click_orchestrate)

# 3. Assemble application workspace layout blocks
control_panel = widgets.VBox([
    prompt_widget,
    widgets.HBox([ratio_widget, scale_widget]),
    widgets.HBox([steps_widget, seed_widget]),
    widgets.Label(value=""),
    generate_btn,
    output_area
])

display(control_panel)
```
